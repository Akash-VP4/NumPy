## 1. Random numbers in NumPy

```python
#Import random mocule

from numpy import random
```

#### Generating random integer

```python
#Generating random integer

rand_num = random.randint(100)

print(f"Random number: {rand_num}")
```

```python
# Generating random number using same seed value

random.seed(42)
for i in range(10):
    print(f"index : {i}, value: {random.randint(10)}")

print("\n------------------------------------------\n")

random.seed(42)
for i in range(10):
    print(f"index : {i}, value: {random.randint(10)}")

print("\n------------------------------------------\n")

for i in range(10):
    random.seed(42)
    print(f"index : {i}, value: {random.randint(10)}")
```

#### Generating random float number

```python
#Generating random float

rand_float = random.rand()

print(f"Random float: {rand_float}")

```

#### Generate random array

```python
rand_arr = random.randint(100,size = (3,5))

print(f"Random array:\n {rand_arr}")

print("\n-----------------------------------\n")

rand_floatarr = random.rand(5)
print(f"Random float array: {rand_floatarr}")
```

```python
rand_floatarr = random.rand(2,3,2)
print(f"Random float array:\n {rand_floatarr}")
```

#### Generate random number from array

```python
rand_num = random.choice([2,4,6,8,10])

print(f"Random number generated from array: {rand_num}")
```

```python
rand_val = random.choice(["apple","banana","orange"])

print(f"Random value: {rand_val}")
```

```python
# Create 2D and 3D array from the given arry

rand_array = random.choice([1,2,3,4,5,6,7,8],size=(2,3))
print(f"Random created 2D array:\n {rand_array}")

print("\n-----------------------------------------------\n")

rand_array = random.choice([1, 2, 3, 4, 5, 6, 7, 8], size=(4,3,3))
print(f"Random created 2D array:\n {rand_array}")
```

## 2. Random data distribution

```python
#Importing numpy random

from numpy import random
```

#### With probability density

```python
#With probability array- The sum of array should be 1

num = random.choice([1,2,3,4,5],p= [0.1,0.5,0.3,0,0.1], size=10)

print(f"Array with probability function:\n {num}")
```

```python
# With probability array- The sum of array should be 1

num = random.choice([1, 2, 3, 4, 5], p=[0.1, 0.5, 0.3, 0, 0.1], size=(2,4))

print(f"Array with probability function:\n {num}")
```

## 5. Normal (Gaussian) Distribution

```python
#Importing random

from numpy import random

```

```python
val = random.normal(size=(5, 6))

print(val)
```

```python
val = random.normal(loc=5,scale=2,size=(3,4))

print(val)
```

#### Visualization of normal distribution

```python
sns.displot(random.normal(loc=0,scale=2,size=10), kind="kde")

plt.show()
```

```python
sns.displot(random.normal(loc=0, scale=1,size=10), kind="kde")

plt.show()
```

## 6. Binomial Distribution

##### -> Normal distribution is continues and the binomial is descrete

```python
#Importing random

from numpy import random
```

```python
rand_num = random.binomial(n=10,p=0.5,size=10)

print(f"Distribution : {rand_num}")
```

#### Visualization of binomial distribution

```python
sns.displot(random.binomial(n=10,p=0.5,size=100))

plt.show()
```

#### Normal and Binomial Distribution

```python

data = {
  'normal':random.normal(loc=5,scale=2,size=100),
  'binomial': random.binomial(n=10,p=0.5,size=100)
}

sns.displot(data,kind='kde')

plt.show()
```
